```python
#horse best past performance

def best_perf(races):
    
    def past(races, when, by, bid, dist):
        #past_races = races.copy()
        past_race = races.query("date < '{}' and {} == '{}' and dist == '{}'".format(when, by, bid, dist))
        if len(past_race) >= 1:
            return past_race.finish_time_micro_sec.min() 
        else:
            return 0
        
    #past(races, races.loc[50,"date"], "horse_id", races.loc[50, "horse_id"], races.loc[50, "dist"])
    races2 = races.copy()
    races2.loc[:, 'max_perf'] = races.apply(lambda x: past(races, x['date'], "horse_id", x['horse_id'], x['dist']), axis=1, result_type="expand")
    #display(races.merge(races.groupby('horse_id')['dist_per_sec'].max(), how="left", on="horse_id"))
    return races2


#display(races.columns)
#best_perf(races)

```

```python
#horse jockey change

def jockey_change(races):
    def past_jockey(races_df, date, hid, cur_jockey):
        races_df2 = races_df.copy()
        races_df2 = races_df2.query("date < '{}' and horse_id == '{}'".format(date, hid)).sort_values("date", ascending=False)['jockey']
        #if len(races_df2.tolist()) >= 1:
            #print(cur_jockey, races_df2.tolist()[0])
        if len(races_df2.tolist()) >= 1 and cur_jockey != races_df2.tolist()[0]:
            return 1
        else:
            return 0
    #races.query("")
    
    #display(races.apply(lambda x: past_jockey(races, x['date'], x['horse_id'], x['jockey'] ), axis=1, result_type="expand" ))
    races2 = races.copy()
    races2["jockey_change"] = races.apply(lambda x: past_jockey(races, x['date'], x['horse_id'], x['jockey'] ), axis=1, result_type="expand" )
    #display(races.set_index("date").rolling(window=2).mean())
    return races2
    



#races = races.sort_values("date")
#races.rolling('2s')['rating'].sum()

#display(races.head())
#races = races.reset_index()
#display(races.head())
#jockey_change(races)
```

```python
def merge_profile(race, profile):
    r = race.copy()
    r = races.merge(profile, how="left", on="horse_id")
    del r["trainer_y"]
    del r["trainer_url_y"]
    del r["trainer_id_y"]
    r = r.rename(columns={"trainer_x": "trainer", "trainer_url_x": "trainer_url", "trainer_id_x": "trainer_id"})
    return r
    
    
    
#merge_profile(races, profiles)
#trainer_x, trainer_url_x, trainer_id_x, 
```

```python

def jockey_place(profile, race):

    #races.iloc[:5, :10]
    df = race.groupby("jockey_id")["pla"].value_counts().unstack(fill_value=0)
    df.columns = ["jockey_place_" + f for f in df.columns]
    #display(df)
    df2 = profile.copy()
    df2 = df2.merge(df, how="left", left_index=True, right_index=True)
    df2[df.columns] = df2[df.columns].fillna(0)
    return df2
    
#jockey_place(profiles, races)
#races.columns
```

```python

def trainer_place(profile, race):
    df = races.groupby("trainer_id")["pla"].value_counts().unstack(fill_value=0)
    df.columns = ["jockey_place_" + c for c in df.columns]
    df2 = profile.copy()
    df2 = df2.merge(df, how="left", left_index=True, right_index=True)
    df2[df.columns] = df2[df.columns].fillna(0)
    return df2

#trainer_place(profiles, races)
```

```python
#display(races.columns)

def past_returns(races_df):
    def past_wins(race, when, hid, win_odds):
        h_race = race.query("date < '{}' and horse_id == '{}'".format(when, hid ))
        #display(len(h_race))
        if len(h_race) > 0:
            return win_odds * (len(h_race.query("pla == '01'")) / len(h_race))
        else:
            return 0
    
    races_df2 = races_df.copy()
    races_df2['win_return'] = races_df.apply(lambda x: past_wins(races_df, x['date'], x['horse_id'], x['win_odds']), axis=1, result_type="expand")
    #display(race_df2 * win_odds)
    #return past_wins(races_df, races_df.loc[40, 'date'], races_df.loc[40, 'horse_id'], races_df.loc[40, 'win_odds'] )
    #race_df.merge(race_df2, how="left", on=)
    return races_df2
    
#%timeit -n 1 -r 1 past_returns(races) #.query("win_return > 1"))
#races.groupby("horse_id")['win_odds'].mean()
```

```python
def past_returns2(races_df):
    def past_wins(race, when, hid, win_odds):
        h_race = race.query("date < '{}' and horse_id == '{}'".format(when, hid ))
        if len(h_race) > 0:
            h_wins = h_race.query("pla == '01'")
            return sum(h_wins['win_odds']) / len(h_race)
            #return win_odds * (len(h_race.query("pla == '01'")) / len(h_race))
        else:
            return 0
    
    race_df2 = races_df.copy()
    race_df2['win_return2'] = races_df.apply(lambda x: past_wins(races_df, x['date'], x['horse_id'], x['win_odds']), axis=1, result_type="expand")
    return race_df2
```

```python
def past_trainer_returns(races_df):
    def past_wins(races, when, jid, win_odds):
        j_races = races.query("date < '{}' and trainer == '{}'".format(when, jid))
        if len(j_races) > 0:
            #j_wins = j_races.query("pla == '01'")
            #res = sum(j_wins['win_odds']) / len(j_races)
            #return res
            return win_odds * (len(j_races.query("pla == '01'")) / len(j_races))
        else:
            return 0
        
        
    races_df2 = races_df.copy()
    #display(races_df2)
    races_df2["trainer_win_return"] = races_df.apply(lambda x: past_wins(races_df, x["date"], x['trainer'], x['win_odds']), axis=1, result_type="expand")
    #display(races_df2["trainer_win_return"])
    return races_df2
  
races['trainer'] = races['trainer'].str.replace("'", "")
#temp_return = past_trainer_returns(races)
```

```python
def convert_dr_to_numeric(races_df):
    races2_df = races_df.copy()
    races2_df["dr"] = pd.to_numeric(races['dr'],errors='coerce')
    return races2_df
```

```python
def convert_to_dummies(races_df, feature):
    return races_df.merge(pd.get_dummies(races_df[feature], prefix=feature), how="left", left_index=True, right_index=True)
    
    
#convert_to_dummies(races, 'pla')
```

```python
def remove_features(races_df, features):
    races_df2 = races_df.copy() # modify races_df should be faster, but copying local var is safer
    for f in features:
        del races_df2[f]
    #display(races_df2.columns)
    return races_df2
    
#remove_features(races, ['race_index', 'race_index_url','rtg', 'trainer', 'jockey', 'jockey_url', 'running_position', 'finish_time', 'gear'])
```

```python
def process_lbw(lbw):
    lbw_dict = {"NOSE":0.05, "N":0.05, "SH":0.1, "HD":0.2, "SN":0.25 ,"DH":0.01, "ML":float("nan"), "TO":float("nan"), "---":0,"HL": 0.5, "1/2": 0.5, "1/4": 0.25, "3/4": 0.75 }
    lbw_dict2 = {"1/2":0.5, "1/4":0.25, "3/4": 0.75}
    #lbw_split = lbw.apply(lambda x: x.split("-"))
    #print(lbw_split)
    def process_lbw_row(lbw_single):
        lbw_single = lbw_single.strip()
        #print(lbw_single)
        if lbw_single == "-":
            return 0
        lbw_split = lbw_single.split("-")
        if len(lbw_split) < 2:
            if lbw_split[0] in lbw_dict.keys():
                return lbw_dict[lbw_split[0]]
            else:
                return int(lbw_split[0])
        elif len(lbw_split) > 2:
            return 0
        else:
            #print(lbw_single[0], lbw_single)
            #print(int(lbw_single[0]))
            #print(int(lbw_dict2[lbw_single[1]]))
            #print(lbw_single)
            return int(lbw_split[0]) + int(lbw_dict2[lbw_split[1]])
    lbw = lbw.str.strip("+")
    return lbw.apply(lambda x: process_lbw_row(x))

#display(race_placements[(race_placements["LBW"] == "+NOSE")]["LBW"].str.strip("+"))
#display(process_lbw(race_placements["LBW"].str.strip("+")))

#display(process_lbw(race_placements["LBW"]))

def process_finish_time(finish_time):
    return finish_time.str.extract(r"(\d+)\.(\d+)\.(\d+)$").apply(lambda x: pd.to_numeric(x[0])*60*100 + pd.to_numeric(x[1])*100 + pd.to_numeric(x[2]), axis = 1)

#display(process_finish_time(race_placements["finish_time"]))
#display(race_placements["finish_time"])


places_dict = {"DISQ": "Disqualified",
               "DNF": "Did not Finish",
               "FE": "Horse fell in the race",
               "PU": "Pulled Up",
               "TNP": "Took no part",
               "UR": "Unseated Rider",
               "VOID": "Void Race", 
               "WR": "Withdrawn - by Starters",
               "WV": "Withdrawn-on Veterinary Ground",
               "WV-A": "Withdrawn-on Vet Grounds + R fee",
               "WX": "Withdrawn-by order of Stewards",
               "WX-A": "withdrawn-by Stewards + R fee"
              }
    
    
def clean_races(race):
    r = race.copy()
    
    r['horse_no'] = pd.to_numeric(race["horse_no"], downcast="signed")
    r["horse_id"] = race["horse"].str.extract(r"\((.+)\)$")[0]
    r["horse"] = race["horse"].str.extract(r"([^(]*)")[0]
    r["actual_wt"] = pd.to_numeric(race["actual_wt"], downcast="unsigned", errors="coerce")
    r["declare_horse_wt"] = pd.to_numeric(race["declare_horse_wt"], downcast="unsigned", errors="coerce")
    r["draw"] = pd.to_numeric(race["draw"], errors="coerce", downcast="unsigned")
    running_position = race.loc[:, "running_position"].str.split(" ")
    r["running_position1"] = running_position.apply(lambda x: int(x[0]) if len(x) >= 2 else 0)
    r["running_position2"] = running_position.apply(lambda x: int(x[1]) if len(x) >= 2 else 0)
    r["running_position3"] = running_position.apply(lambda x: int(x[2]) if len(x) >= 3 else 0)
    r["running_position4"] = running_position.apply(lambda x: int(x[3]) if len(x) >= 4 else 0)
    r["win_odds"] = pd.to_numeric(race["win_odds"], errors="coerce", downcast="unsigned")
    r["race_number"] = pd.to_numeric(race["race_number"], errors="corece")
    r["race_date"] = pd.to_datetime(race["race_date"])
    r['finish_time'] = process_finish_time(race["finish_time"])
    r['LBW'] = process_lbw(race["LBW"])
    r['jockey'] = race.loc[:, "jockey"].str.replace("'", "")
    r['trainer'] = race.loc[:, "trainer"].str.replace("'", "")
    r['pla'] = race['pla'].str.replace(r"(\d+) DH", r"\1")
    #r['owner'] = race.loc[:, "owner"].str.replace("'", "")
    for k in ['running_position', 'horse_url', 'jockey_url', 'trainer_url', 'race_url']:
        del r[k]
    
    return r

#clean_races(race_placements).loc[:, :]

clean_races(race_placements).to_pickle("data/race_placements_clean.pickle")
```

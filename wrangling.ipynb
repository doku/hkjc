```python
def clean_profile(profile):
    prof = profile.copy()
    #print(prof)
    prof["name_name"] = profile.loc[:,"name"].str.extract(r"(.*)\(.{4}\)$")
    country_age = profile.loc[:, "country_age"].str.split("/")
    prof["country"] = country_age.apply(lambda x: x[0].strip())
    prof["age"] = pd.to_numeric(country_age.apply(lambda x: x[1].strip()))
    colour_sex = profile.loc[:,"colour_sex"].str.split("/")
    prof["sex"] = colour_sex.apply(lambda x: x[-1].strip())
    colour = colour_sex.apply(lambda x: [f.strip() for f in x[:-1]])
    prof["colour_one"] = colour.apply(lambda x: x[0] if len(x) >= 1 else "")
    prof["colour_two"] = colour.apply(lambda x: x[1] if len(x) >= 2 else "")
    prof["season_stakes"] = profile.loc[:, "season_stakes"].str.extract(r"^\$(.*)")[0].str.replace(",", "").astype("int")
    prof["total_stakes"] = profile.loc[:, "total_stakes"].str.extract(r"^\$(.*)")[0].str.replace(",", "").astype("int")
    no_123_starts = profile.loc[:, "no_123_starts"].str.split("-")
    prof["no_123_starts_1"] = no_123_starts.apply(lambda x: x[0]).astype(int)
    prof["no_123_starts_2"] = no_123_starts.apply(lambda x: x[1]).astype(int)
    prof["no_123_starts_3"] = no_123_starts.apply(lambda x: x[2]).astype(int)
    prof["no_123_starts_4"] = no_123_starts.apply(lambda x: x[3]).astype(int)
    stable_location = profile.loc[:, "stable_location"].str.split("\n")
    prof["stable_location"] = stable_location.apply(lambda x: x[0])
    prof["stable_location_date"] = stable_location.apply(lambda x: x[1] if len(x) == 2 else "")
    prof["current_rating"] = pd.to_numeric(profile.loc[:, "current_rating"], downcast="integer")
    prof["start_season_rating"] = pd.to_numeric(profile.loc[:, "start_season_rating"], downcast="integer")
    prof["trainer_id"] = profile.loc[:, "trainer_url"].str.extract(r"=(.*)$")
    prof["sire_id"] = profiles.loc[:, "sire_url"].str.extract(r"/([^/]+)$")
    prof["no_start_past_10"] = pd.to_numeric(profiles['no_start_past_10'])
    prof['owner'] = profiles['owner'].str.replace("'", "").str.replace(" ", ".")

    #print(stable_location.apply(lambda x: x[1]))
    #print(stable_location)
    #print(profile.loc[:, "stable_location"])
    
    
    return prof

#clean_profile(profiles).to_pickle("data/profiles.pickle")
clean_profile(profiles)   
    
    
```

```python

def process_lbw(lbw):
    lbw_dict = {"NOSE":0.05, "N":0.05, "SH":0.1, "HD":0.2, "SN":0.25 ,"DH":0.01, "ML":float("nan"), "TO":float("nan"), "---":0,"HL": 0.5, "1/2": 0.5, "1/4": 0.25, "3/4": 0.75 }
    lbw_dict2 = {"1/2":0.5, "1/4":0.25, "3/4": 0.75}
    #lbw_split = lbw.apply(lambda x: x.split("-"))
    #print(lbw_split)
    def process_lbw_row(lbw_single):
        lbw_single = lbw_single.strip()
        #print(lbw_single)
        if lbw_single == "-":
            return 0
        lbw_single = lbw_single.split("-")
        if len(lbw_single) < 2:
            if lbw_single[0] in lbw_dict.keys():
                return lbw_dict[lbw_single[0]]
            else:
                return int(lbw_single[0])
        elif len(lbw_single) > 2:
            return 0
        else:
            #print(lbw_single[0], lbw_single)
            #print(int(lbw_single[0]))
            #print(int(lbw_dict2[lbw_single[1]]))
            #print(lbw_single)
            return int(lbw_single[0]) + int(lbw_dict2[lbw_single[1]])
    return lbw.apply(lambda x: process_lbw_row(x))

def process_finish_time(finish_time):
    return finish_time.str.extract(r"(\d+)\.(\d+)\.(\d+)$").apply(lambda x: pd.to_numeric(x[0])*60*100 + pd.to_numeric(x[1])*100 + pd.to_numeric(x[2]), axis = 1)


gear_dict = {"B":"Blinkers", 
             "BO": "Blinkers_Cowl", 
             "CC": "Cornell_Collar",
             "XB":"Crossed_Nose_Band",
             "E": "Ear_Plugs",
             "H": "Hood",
             "P": "Pacifier",
             "PS": "Pacifer_with_one_Cowl",
             "PC": "Pacifier_with_Cowls",
             "SR": "Shadow_Roll",
             "SB": "Sheepskin_Browband",
             "CP": "Sheepskn_Cheek_Pieces",
             "CO": "Sheepskin_Cheek_Piece_One_Side",
             "TT": "Tongue_Tie",
             "V": "Visor"
            }
gear_times = {
             "0": "gear_on",
             "1": "first_time",
             "2": "replace",
             "-": "removed"
            }

places_dict = {"DISQ": "Disqualified",
               "DNF": "Did not Finish",
               "FE": "Horse fell in the race",
               "PU": "Pulled Up",
               "TNP": "Took no part",
               "UR": "Unseated Rider",
               "VOID": "Void Race", 
               "WR": "Withdrawn - by Starters",
               "WV": "Withdrawn-on Veterinary Ground",
               "WV-A": "Withdrawn-on Vet Grounds + R fee",
               "WX": "Withdrawn-by order of Stewards",
               "WX-A": "withdrawn-by Stewards + R fee"
              }
    
    

def split_gear(gear_items):
    res = {}
    for gear_item in gear_items:
        if gear_item == "--":
            return res
        if gear_item[-1] in ["1","2", "-"]:
            res[gear_item[:-1]] = gear_times[gear_item[-1]]
            #return (gear_item[:-1], gear_times[gear_item[-1]]) 
        else:
            res[gear_item] = gear_times["0"]
            #return (gear_item, 0)
    return res

    #gear_list = gear.str.split("/").apply(lambda x: split_gear(x))
    #res = pd.DataFrame()
    #for gi, gg in gear_dict.items():
        #print(gi, gg)
    #    res["gear_"+ gg] = gear_list.apply(lambda x: x[gi] if gi in x.keys() else "gear_off")
    #return res


def clean_races(race):
    r = race.copy()
    rc_track_course = race.loc[:, "rc_track_course"].str.split("/")
    r['race_course'] = rc_track_course.apply(lambda x: x[0].strip())
    r['race_course_2'] = rc_track_course.apply(lambda x: x[1].strip() if len(x) >= 2 else "")
    r['race_course_3'] = rc_track_course.apply(lambda x: x[2].strip() if len(x) >= 3 else "").str.strip("\"")
    #print(race.loc[80:120, "lbw"])
    r['lbw'] = process_lbw(race.loc[:, "lbw"])
    running_position = race.loc[:, "running_position"].str.split(" ")
    r['running_position_1'] = running_position.apply(lambda x: int(x[0]) if len(x) > 2 else 0)
    r['running_position_2'] = running_position.apply(lambda x: int(x[1]) if len(x) > 2 else 0)
    r['running_position_3'] = running_position.apply(lambda x: int(x[2]) if len(x) > 2 else 0)
    r['running_position_4'] = running_position.apply(lambda x: int(x[3]) if len(x) == 4 else 0)
    
    r['finish_time_micro_sec'] = process_finish_time(race.loc[:, "finish_time"])
    r['date'] = pd.to_datetime(race.loc[:, "date"])
    r['dist'] = pd.to_numeric(race.loc[:, "dist"])
    r['rating'] = pd.to_numeric(race.loc[:, "rtg"], errors="coerce")
    r['win_odds'] = pd.to_numeric(race.loc[:, "win_odds"], errors="coerce")
    r['act_wt'] = pd.to_numeric(race.loc[:, "act_wt"])
    r['declare_horse_wt'] = pd.to_numeric(race.loc[:, "declare_horse_wt"], errors="coerce")
    gear_list = race.loc[:, "gear"].str.split("/").apply(lambda x: split_gear(x))
    for gi, gg in gear_dict.items():
        #print(gi, gg)
        r["gear_"+ gg] = gear_list.apply(lambda x: x[gi] if gi in x.keys() else "gear_off")
 
    r['dist_per_sec'] = (r.loc[:, "dist"] / r.loc[:, "finish_time_micro_sec"])*100
    
    r['jockey_id'] = race.jockey_url.str.extract(r"JockeyId=([^&]+)&Season=Current$")
    r['trainer_id'] = race.trainer_url.str.extract(r"TrainerId=([^&]+)&Season=Current$")
    r['jockey'] = race['jockey'].str.replace("'", "")
    r['trainer'] = race['trainer'].str.replace("'", "")
    r['trainer'] = race['trainer'].str.replace("'", "").str.replace(" ", ".")
    r['jockey'] = race['jockey'].str.replace("'", "").str.replace(" ", ".")
    r['race_course_3'] = race['race_course_3'].str.replace('"', "")
    r['owner'] = race['owner'].str.replace("'", "").str.replace(" ", ".")
    #print(race.loc[:, "g"].unique())



    #print(process_gear(race.loc[:, "gear"]))
    
    #lbw_dict = {"NOSE":0.05, "N":0.05, "SH":0.1, "HD":0.2, "SN":0.25 ,"DH":0.01, "ML":float("nan"), "TO":float("nan"), "---":0,"HL": 0.5 }
    #lbw_split = race.loc[:, "lbw"].apply(lambda x: x.split("-"))
    #print(lbw_split)
    
    #print(race.loc[:, "lbw"])
    #r['turf'] = rc_track_course.apply(lambda x: x[1].strip())
    #print(race.loc[:, "running_position"].str.split(" ").apply(lambda x: x))
    
    #print(race.loc[:, "rc_track_course"].nunique())
    #print(rc_track_course.apply(lambda x: len(x)))
    
    #r['course'] = rc_track_course.apply(lambda x: x[2].strip())
    #print(rc_track_course)
    
    
    return r


    
clean_races(races)
```
